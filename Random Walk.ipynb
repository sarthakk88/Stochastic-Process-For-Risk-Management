# Random Walk Simulation

## Random walk without drift

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
def random_walk(p,q,n,start,variance) :
    np.random.seed(5)
    probability_at_each_step = np.random.rand(n)
    position = start 
    position_at_each_step = [position] 
    for i in range(n) :
        if probability_at_each_step[i] <= p:
            position = position + np.random.randn()*variance
        elif probability_at_each_step[i] <= p+q :
            position = position - np.random.randn()*variance 
        position_at_each_step.append(position)
    return position_at_each_step

position_at_each_step = random_walk(0.6,0.3,100,10,2)
plt.xlabel("Time")
plt.ylabel("Value")
plt.plot(np.arange(0,101),position_at_each_step)
```

## Random walk with drift

```python
def random_walk_drift(p,q,n,start,variance,drift) :
    np.random.seed(7)
    probability_at_each_step = np.random.rand(n)
    position = start 
    position_at_each_step = [position + drift] 
    for i in range(n) :
        if probability_at_each_step[i] <= p:
            position = position + np.random.randn()*variance + drift
        elif probability_at_each_step[i] <= p+q :
            position = position - np.random.randn()*variance + drift
        position_at_each_step.append(position)
    return position_at_each_step

position_at_each_step = random_walk_drift(0.6,0.3,100,10,10,4)
plt.xlabel("Time")
plt.ylabel("Value")
plt.plot(np.arange(0,101),position_at_each_step)
```
