# Brownian Motion Simulation

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
def BrownianMotion(t,n):
    dt = t/(n-1)
    X = [0] * n
    dX = [0] * n
    dX[0] = np.sqrt(dt) * np.random.randn()
    X[0] = dX[0]
    for j in range(1, n):
        dX[j] = np.sqrt(dt) * np.random.randn()
        X[j] = X[j-1] + dX[j]
    return X
plt.xlabel("Time")
plt.ylabel("X(t)")
X = BrownianMotion(1.0,501)
plt.plot(X)
```
